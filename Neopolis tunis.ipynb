Écrire un décorateur nommé log_content , qui enregistre chaque variable de façon transparente à la fonction
Exemple:
```
@log_content
def example(a, b, c):
    a = b + c
    b = a + c
    c = a + b
    d = "hi there"
    result = a + b + c
    return result
```

````python
def decorator_with_arguments(function):
    def wrapper_accepting_arguments(arg1, arg2, arg3):
        func = function(arg1, arg2,arg3) 
        func = arg1+arg2+arg3
        return func
    return wrapper_accepting_arguments

@decorator_with_arguments
def exemple (a, b , c):
    r= a+ b+c
    a = b + c
    b = a + c
    c = a + b
    d = "hi there"
    result = a+ b+c 
    return result
````
